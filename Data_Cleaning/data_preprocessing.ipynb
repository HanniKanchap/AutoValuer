```python
import pandas as pd
```

```python
df1 = pd.read_csv("../Data/olx_car_listings_expanded.csv")
df2 = pd.read_csv("../Data/olx_car_listings_expanded1.csv")
```

```python
df = pd.concat([df1, df2], ignore_index=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Information'] = df['Information'].str.split(" - ")
```

```python
df['Information']
```

```python
def fetch_year(data):
    return data[0]

df['Year'] = df['Information'].apply(fetch_year)
df['Year'] = df['Year'].astype(int)
```

```python
df.head()
```

```python
def fetch_distance(data):
    return data[1]

df['Distance Covered'] = df['Information'].apply(fetch_distance)
```

```python
def filter_dis(data):
    return data.replace('km',"")

df['Distance Covered'] = df['Distance Covered'].apply(filter_dis)
```

```python
df['Distance Covered'] = df['Distance Covered'].str.strip()
df['Distance Covered'] = df['Distance Covered'].str.replace(',','').astype(float)
```

```python
df.head()
```

```python
companies = ['maruti suzuki', 'hyundai', 'honda', 'toyota', 'tata', 'mahindra', 'mercedes benz','ford','volkswagen','audi','nissan' ,'bmw', 'kia','jeep','mg','renault','skoda','mini cooper','jaguar','land rover','datsun','chevrolet']

def filter_brand(curr):
    for com in companies:
        curr = curr.replace('-',' ')
        if com in curr.lower():
            return com
        
    return 'Unknown'

df['Brand_Title'] = df['Title'].apply(filter_brand)

```

```python
df['Brand'] = df.apply(
    lambda row: row['Brand_Title'] if row['Brand_Title'] != 'Unknown' else row['Brand'],
    axis=1
)
```

```python
df.head()
```

```python
df.drop(['Brand_Title','Information'],axis = 1,inplace= True)
```

```python
def filter_price(data):
    p =  data.replace('₹ ','')
    p = p.replace(',','')
    return int(p)

df['Price'] = df['Price'].apply(filter_price)
```

```python
import pandas as pd
import re
from datetime import datetime, timedelta

# Expanded regex pattern
date_pattern = r"(Today|Yesterday|\d{1,2} days ago|\d{1,2} [A-Za-z]{3})$"

# Extract date string
df["Date_Posted"] = df["Location"].str.extract(date_pattern)

# Clean location by removing the date suffix
df["Clean_Location"] = df["Location"].str.replace(date_pattern, "", regex=True).str.strip()

# Convert relative and absolute dates
def parse_date(date_str):
    if pd.isna(date_str):
        return pd.NaT
    if date_str == "Today":
        return datetime.today()
    if date_str == "Yesterday":
        return datetime.today() - timedelta(days=1)
    if "days ago" in date_str:
        days = int(re.search(r"(\d+)", date_str).group(1))
        return datetime.today() - timedelta(days=days)
    try:
        return datetime.strptime(date_str, "%d %b")
    except:
        return pd.NaT

# Apply conversion
df["Date_Posted"] = df["Date_Posted"].apply(parse_date)
```

```python
df['Location'] = df['Clean_Location']
```

```python
df.drop(['Clean_Location','Date_Posted'],axis = 1,inplace=True)
```
