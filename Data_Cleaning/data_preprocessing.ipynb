```python
import pandas as pd
```

```python
df1 = pd.read_csv("../Data/olx_car_listings_expanded.csv")
df2 = pd.read_csv("../Data/olx_car_listings_expanded1.csv")
```

```python
df1['Information'] = df1['Information'].str.split(" - ")
df2['Information'] = df2['Information'].str.split(" - ")
```

```python
df1['Information']
```

```python
def fetch_year(data):
    return data[0]
df1['Year'] = df1['Information'].apply(fetch_year)
df2['Year'] = df2['Information'].apply(fetch_year)
```

```python
df1.head()
```

```python
df2.head()
```

```python
def fetch_distance(data):
    return data[1]
df1['Distance Covered'] = df1['Information'].apply(fetch_distance)
df2['Distance Covered'] = df2['Information'].apply(fetch_distance)
```

```python
df1['Year'] = df1['Year'].astype(int)
df2['Year'] = df2['Year'].astype(int)
```

```python
def filter_dis(data):
    return data.replace('km',"")

df2['Distance Covered'] = df2['Distance Covered'].apply(filter_dis)
df1['Distance Covered'] = df1['Distance Covered'].apply(filter_dis)
```

```python
df1['Distance Covered'] = df1['Distance Covered'].str.strip()
df1['Distance Covered'] = df1['Distance Covered'].str.replace(',','').astype(float)
```

```python
df2['Distance Covered'] = df2['Distance Covered'].str.strip()
df2['Distance Covered'] = df2['Distance Covered'].str.replace(',','').astype(float)
```

```python
df1.head()
```

```python
df2.head()
```

```python
companies = ['maruti suzuki', 'hyundai', 'honda', 'toyota', 'tata', 'mahindra', 'mercedes benz','ford','volkswagen','audi','nissan' ,'bmw', 'kia','jeep','mg','renault']
def filter_brand(curr):
    for com in companies:
        if com in curr.lower():
            return com
        
    return 'Unknown'

df1['Brand_Title'] = df1['Title'].apply(filter_brand)

```

```python
df2['Brand_Iitle'] = df2['Title'].apply(filter_brand)
```

```python
df1['Brand'] = df1.apply(
    lambda row: row['Brand_Title'] if row['Brand_Title'] != 'Unknown' else row['Brand'],
    axis=1
)
```

```python
df2['Brand'] = df2.apply(
    lambda row: row['Brand_Iitle'] if row['Brand_Iitle'] != 'Unknown' else row['Brand'],
    axis=1)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df2.drop(['Brand_Iitle'],axis = 1,inplace= True)
df1.drop(['Brand_Title'],axis = 1,inplace=True)
```

```python
df2.drop(['Information'],axis = 1,inplace= True)
df1.drop(['Information'],axis = 1,inplace=True)
```

```python
def filter_price(data):
    p =  data.replace('₹ ','')
    p = p.replace(',','')
    return int(p)

df2['Price'] = df2['Price'].apply(filter_price)
df1['Price'] = df1['Price'].apply(filter_price)
```

```python
combined_df = pd.concat([df1, df2], ignore_index=True)
```
